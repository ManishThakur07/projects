```python
from sklearn.datasets import load_wine
```

```python
df = load_wine()
```

```python
df
```

```python
x = df["data"]
y = df["target"]
x.shape
```

```python
from sklearn.decomposition import PCA
P = PCA(n_components=5)
pp = P.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(pp,y,test_size=0.2,random_state=15)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
predc = dtc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test,predc)
```

```python
cmd = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=df["target_names"])
cmd.plot()
```
