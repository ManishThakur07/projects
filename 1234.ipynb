```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("car_price.csv")
df
```

```python
df = df.drop(columns = ["Unnamed: 0"])
df
```

```python
le = LabelEncoder()
df["transmission"] = le.fit_transform(df["transmission"])
df["fuel_type"] = le.fit_transform(df["fuel_type"])
df["ownership"] = le.fit_transform(df["ownership"])
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
a = df["car_name"]
cv = CountVectorizer()
cc = pd.DataFrame()
c = pd.DataFrame(cv.fit_transform(df["car_name"].fillna('')).toarray())
cc = pd.concat([cc,c],axis = 1)
```

```python
cc
```

```python
df["car_prices_in_rupee"] = df["car_prices_in_rupee"].str.replace("Lakh"," ",regex = False)
df["kms_driven"] = df["kms_driven"].str.replace("kms"," ",regex = False)
# df["ownership"] = df["ownership"].str.replace("Lakh"," ",regex = False)
df["engine"] = df["engine"].str.replace("cc"," ",regex = False)
df["Seats"] = df["Seats"].str.replace("Seats"," ",regex = False)
```

```python
y = df.iloc[:,1]
y
```

```python
from sklearn.decomposition import PCA
a = PCA(n_components=1)
b = a.fit_transform(cc) 
b.shape
```

```python
xx = df.drop(columns=["car_name","car_prices_in_rupee"])
```

```python
datf = pd.DataFrame(b)
```

```python
x = pd.concat([datf,xx],axis = 1)
```

```python
x.rename(columns={0: "car_name"}, inplace=True)
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```
